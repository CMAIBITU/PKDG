```python
def extract_amp_spectrum(img_np):
    # trg_img is of dimention CxHxW (C = 3 for RGB image and 1 for slice)
    # ftt (3,384,384)
    # 快速傅里叶变换算法得到频率分布
    fft = np.fft.fft2( img_np, axes=(-2, -1) )
    # np 操作中提取对应振幅谱（低级语义信息），相位谱（高级信息）
    amp_np, pha_np = np.abs(fft), np.angle(fft)
    return amp_np
i=501
 # original=input('请输入原始数据所在的路径：')
original = r"/home/zll/Fundus/ORIGA/test/image/"
os.chdir(original)
# 用列表推导式，遍历文件夹下所有文件名放在file内，如列表推导式看不习惯，请用下面注释掉的5句。
filename_no_suffix_list=[ file for file in os.listdir() if file.endswith('.jpg')] # 如果后缀为.xlsm
filename_no_suffix_list.sort()
for client_idx in filename_no_suffix_list:
    each_data_path = original+client_idx
    print(each_data_path)
    # 读取到的img (384,384,3)
    img = Image.open(each_data_path).convert("RGB")
    # img = np.asarray(img, np.float32)
    img = img.resize((384, 384), Image.BICUBIC)
    img = np.asarray(img)
    print(img.shape)
    # im_trg 将img 图像进行翻转 （3*384*384）
    im_trg = img.transpose((2, 0, 1))
    amp = extract_amp_spectrum(im_trg)
    # 将图像中的低频部分移动到图像的中心，默认是在左上角
    amp_target_shift = np.fft.fftshift(amp, axes=(-2, -1))
    # fft结果是复数, 其绝对值结果是振幅，取对数的目的是将数据变换到0~255
    # 这里用归一化 除上最大值并且限制范围到0，1之间，转化为PIL.Image格式
    image = np.clip((np.log(amp_target_shift) / np.max(np.log(amp_target_shift))).transpose((1, 2, 0)), 0, 1)
    #file_name = 'D:/pythonProject7/FedDG-ELCFS-main/dataset/Fundus/client4/freq_amp_npy/amp_sample{}'.format(client_idx)
    #np.save(file_name, image)

    file_name2='/home/zll/fedDG/dataset/Fundus/client5/freq_amp_npy/amp_sample'+str(i)+'.npy'
    print(file_name2)

    #image.transpose(1, 2, 0)
    im3 = np.asarray(image*255.0)
    # print(im3.shape)
    # print(im3.max())
    #np.save(file_name2, im3)
    # 以文件的原名称存储新数据类型的数据npy
    # 保存npy文件 np.save
    np.save(file_name2, im3)


    #image = np.transpose(im3, (2, 0, 1))
    #image = Image.fromarray(im3,mode='RGB')
    file_name2='/home/zll/fedDG/dataset/Fundus/client5/freq_amp_jpg/amp_sample'+str(i)+'.jpg'
    i = i + 1
    cv2.imwrite(file_name2, im3)
```
